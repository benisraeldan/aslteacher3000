```python
import os
import math
import cv2

# we want 30 fps
# test: if the video is less than 30 fps we take the last frame untill we get to 30
#       if the video is more than 30 fps we reduce frames
```

```python
a = r'c:\Users\gural\Desktop\s';
folderList = os.listdir(a)
for folder in folderList:
    listing = os.listdir(a + '\\' + folder)
    count = 1
    lst = {}
    for file in listing:    
        video = cv2.VideoCapture(a + '\\' + folder + "\\" + file)    
        framerate = video.get(5)
        print(framerate)
        os.makedirs(a + "\\" +folder + "\\" + folder + "_" +  str(int(count)))
        b = a + "\\" + folder + "\\" + folder + "_" +  str(int(count))
        while (video.isOpened()):
            frameId = video.get(1)
            success,image = video.read()         
            try:
                if(image.any() != None):
                    image=cv2.resize(image,(128,128), interpolation = cv2.INTER_AREA)
            except:
                if(image != None):
                    image=cv2.resize(image,(128,128), interpolation = cv2.INTER_AREA)
            if (success != True):
                break

            lst[frameId] = b +"\\image_" + str(int(frameId +1)) + ".jpg"
            filename = lst[frameId]
            print(filename)
            cv2.imwrite(filename,image)
        video.release()
        print('done')
        count+=1
```
